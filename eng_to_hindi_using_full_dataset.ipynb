```python
# Binarize the dataset
!fairseq-preprocess \
    --source-lang en --target-lang hi \
    --trainpref /content/drive/MyDrive/eng_to_hindi/wmt17_en_hi/train --validpref /content/drive/MyDrive/eng_to_hindi/wmt17_en_hi/valid \
    --destdir /content/drive/MyDrive/eng_to_hindi/data-bin/wmt17_en_hi --thresholdtgt 0 --thresholdsrc 0 \
    --workers 20
```

```python
# Train the model
!mkdir -p /content/drive/MyDrive/eng_to_hindi/checkpoints/fconv_wmt_en_hi
!fairseq-train \
    /content/drive/MyDrive/eng_to_hindi/data-bin/wmt17_en_hi \
    --arch fconv_wmt_en_de \
    --dropout 0.2 \
    --criterion label_smoothed_cross_entropy --label-smoothing 0.1 \
    --optimizer nag --clip-norm 0.1 \
    --lr 0.5 --lr-scheduler fixed --force-anneal 50 \
    --max-epoch 10 \
    --max-tokens 4000 \
    --save-dir /content/drive/MyDrive/eng_to_hindi/checkpoints/fconv_wmt_en_hi
```

```python
# loading a trained model
from fairseq.models.fconv import FConvModel
en2hi = FConvModel.from_pretrained(
  '/content/drive/MyDrive/eng_to_hindi/checkpoints/fconv_wmt_en_hi',
  checkpoint_file='checkpoint_best.pt',
  data_name_or_path='/content/drive/MyDrive/eng_to_hindi/data-bin/wmt17_en_hi',
  bpe='subword_nmt',
  bpe_codes='/content/drive/MyDrive/eng_to_hindi/wmt17_en_hi/code'
)
```

## **Testing Script**

```python
source = '/content/drive/MyDrive/eng_to_hindi/demo/eng.txt'
dest = '/content/drive/MyDrive/eng_to_hindi/demo/hindi.txt'

eng_file = open(source, "r")
eng_content = eng_file.read()
print('Contents of English file: ')
print(eng_content)
eng_content = eng_content.split('\n')

hindi_content = ''
for sentence in eng_content:
  hindi_content += en2hi.translate(sentence) + '\n'
hindi_file = open(dest, "w")
hindi_file.write(hindi_content)
print('\nContents of Hindi file: ')
print(hindi_content)

eng_file.close()
hindi_file.close()
```
